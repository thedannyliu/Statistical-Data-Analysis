## 3-dim K means

```python
data = yoxi_used
cols = list(data.columns.values)
cols.pop(cols.index('yoxi_7'))
data = data[cols+['yoxi_7']]
cols = list(data.columns.values)
cols.pop(cols.index('yoxi_14'))
data = data[cols+['yoxi_14']]
cols = list(data.columns.values)
cols.pop(cols.index('yoxi_15'))
data = data[cols+['yoxi_15']]
data.head()
```

```python
X = data
#elbow method
wcss = []
for i in range(1,11):
    k_means = KMeans(n_clusters=i,init='k-means++', random_state=42)
    k_means.fit(X)
    wcss.append(k_means.inertia_)
#plot elbow curve
plt.plot(np.arange(1,11),wcss)
plt.xlabel('Clusters')
plt.ylabel('SSE')
plt.title("SSE track")
plt.show()
# K=3
```

```python
k_means_optimum = KMeans(n_clusters = 3, init = 'k-means++',  random_state=42)
y = k_means_optimum.fit_predict(X)
print(y)
```

```python
data['cluster'] = y  
# the above step adds extra column indicating the cluster number for each country
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
```

```python
kplot = plt.axes(projection='3d')
xline = np.linspace(0, 15, 1000)
yline = np.linspace(0, 5, 1000)
zline = np.linspace(0, 15, 1000)
kplot.plot3D(xline, yline, zline, 'black')
# Data for three-dimensional scattered points
kplot.scatter3D(data1.yoxi_7, data1.yoxi_14, data1.yoxi_15, c='red', label = 'Cluster 1')
kplot.scatter3D(data2.yoxi_7, data2.yoxi_14, data2.yoxi_15, c ='green', label = 'Cluster 2')
kplot.scatter3D(data3.yoxi_7, data3.yoxi_14, data3.yoxi_15, c ='blue', label = 'Cluster 3')
plt.scatter(k_means_optimum.cluster_centers_[:,0], k_means_optimum.cluster_centers_[:,1], color = 'indigo', s = 200)
plt.legend()
plt.title("Kmeans (K=3)")
plt.show()
```

```python
yoxi_time = pd.read_csv('yoxi_2021H1_time_revised.csv')
yoxi_time
```

```python
yoxi_clear_t = yoxi_time.drop(yoxi_time[yoxi_time['yoxi_15']=='0'].index, axis=0)
yoxi_clear_t = yoxi_clear_t.drop(yoxi_clear_t[yoxi_clear_t['yoxi_3']=='1900/1/0'].index, axis=0)
yoxi_clear_t = yoxi_clear_t.drop(yoxi_time[yoxi_time['yoxi_15'].isna()].index, axis=0)
yoxi_clear_t['yoxi_14'] = yoxi_clear_t['yoxi_14'].map({'[]':0, '[1]':1, '[2]':2, '[1,2]':3, '[2,1]':3})
yoxi_clear_t['yoxi_15'] = yoxi_clear_t['yoxi_15'].map({'1-100':50, '101-150':125, '151-200':175, '201-250':225, '251-300':275, '301-400':350, '401-500':450, '500-':500,})
yoxi_clear_t
```

```python
#留 1.行駛分鐘數 2.乘客偏好代碼 3.乘車金額區間
yoxi_used = yoxi_clear_t.drop(['yoxi_0','yoxi_1','yoxi_2','yoxi_3','yoxi_4','yoxi_5','yoxi_6','yoxi_8','yoxi_9','yoxi_10','yoxi_11','yoxi_12','yoxi_13',], axis=1)
yoxi_used
```

### Recency

```python
# '%Y-%m-%d %H:%M:%S.%f'
yoxi_clear_t['yoxi_3'] = pd.to_datetime(yoxi_clear_t['yoxi_3'], format = '%Y/%m/%d')
yoxi_clear_t
```

```python
df_recency = yoxi_clear_t.groupby(by='yoxi_2',
						as_index=False)['yoxi_3'].max()
df_recency.columns = ['CustomerID', 'LastPurchaseDate']
recent_date = df_recency['LastPurchaseDate'].max()
df_recency['Recency'] = df_recency['LastPurchaseDate'].apply(
	lambda x: (recent_date - x).days)
df_recency.head(20)
```

### Monetary

```python
monetary_df = yoxi_clear_t.groupby(by='yoxi_2', as_index=False)['yoxi_15'].sum()
monetary_df.columns = ['CustomerID', 'Monetary']
monetary_df.head(20)
```

### Frequency

```python
frequency_df = yoxi_clear_t.drop_duplicates().groupby(
    by=['yoxi_2'], as_index=False)['yoxi_3'].count()
frequency_df.columns = ['CustomerID', 'Frequency']
frequency_df.head(20)
```

```python
rf_df = df_recency.merge(frequency_df, on='CustomerID')
rfm_df = rf_df.merge(monetary_df, on='CustomerID').drop(
	columns='LastPurchaseDate')
rfm_df.head(20)
```

```python
rfm_df['R_rank'] = rfm_df['Recency'].rank(ascending=False)
rfm_df['F_rank'] = rfm_df['Frequency'].rank(ascending=True)
rfm_df['M_rank'] = rfm_df['Monetary'].rank(ascending=True)

# normalizing the rank of the customers
rfm_df['R_rank_norm'] = (rfm_df['R_rank']/rfm_df['R_rank'].max())*100
rfm_df['F_rank_norm'] = (rfm_df['F_rank']/rfm_df['F_rank'].max())*100
rfm_df['M_rank_norm'] = (rfm_df['F_rank']/rfm_df['M_rank'].max())*100

# rfm_df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)

rfm_df.head(30)
```

```python
rfm_df['RFM_Score'] = 0.15*rfm_df['R_rank_norm']+0.25 * \
    rfm_df['F_rank_norm']+0.6*rfm_df['M_rank_norm']
rfm_df['RFM_Score'] *= 0.05
rfm_df = rfm_df.round(2)
rfm_df[['CustomerID', 'RFM_Score']].head(20)
```

```python
rfm_df["Customer_segment"] = np.where(rfm_df['RFM_Score'] >
                                      4.5, "Top Customers",
                                      (np.where(
                                        rfm_df['RFM_Score'] > 4,
                                        "High value Customer",
                                        (np.where(
    rfm_df['RFM_Score'] > 3,
                             "Medium Value Customer",
                             np.where(rfm_df['RFM_Score'] > 1.6,
                            'Low Value Customers', 'Lost Customers'))))))
rfm_df[['CustomerID', 'RFM_Score', 'Customer_segment']].head(30)
```

```python
plt.pie(rfm_df.Customer_segment.value_counts(),
        labels=rfm_df.Customer_segment.value_counts().index,
        autopct='%.0f%%')
plt.show()
```
